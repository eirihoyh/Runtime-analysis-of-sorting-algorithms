# Tables of Sorted lists

```python
import pandas as pd
merge = pd.read_pickle('../data/benchmark_merge_sort')
merge['Algorithm'] = 'Merge sort'
tim = pd.read_pickle('../data/benchmark_tim_sort')
tim['Algorithm'] = 'Tim sort'
quick = pd.read_pickle('../data/benchmark_quick_sort')
quick['Algorithm'] = 'Quick sort'
insertion = pd.read_pickle('../data/benchmark_insertion_sort')
insertion['Algorithm'] = 'Insertion sort'
bubble = pd.read_pickle('../data/benchmark_bubble_sort')
bubble['Algorithm'] = 'Bubble sort'
numpy = pd.read_pickle('../data/benchmark_numpy_sort')
numpy['Algorithm'] = 'Numpy sort'
python = pd.read_pickle('../data/benchmark_python_sort')
python['Algorithm'] = 'Pyhton sort'
```

```python
merge_increasing = pd.DataFrame(merge.query("list_type == 'increasing_list'"))
tim_increasing = pd.DataFrame(tim.query("list_type == 'increasing_list'"))
quick_increasing = pd.DataFrame(quick.query("list_type == 'increasing_list'"))
insertion_increasing = pd.DataFrame(insertion.query("list_type == 'increasing_list'"))
bubble_increasing = pd.DataFrame(bubble.query("list_type == 'increasing_list'"))
numpy_increasing = pd.DataFrame(numpy.query("list_type == 'increasing_list'"))
python_increasing = pd.DataFrame(python.query("list_type == 'increasing_list'"))
```

```python
merge_decreasing = pd.DataFrame(merge.query("list_type == 'decreasing_list'"))
tim_decreasing = pd.DataFrame(tim.query("list_type == 'decreasing_list'"))
quick_decreasing = pd.DataFrame(quick.query("list_type == 'decreasing_list'"))
insertion_decreasing = pd.DataFrame(insertion.query("list_type == 'decreasing_list'"))
bubble_decreasing = pd.DataFrame(bubble.query("list_type == 'decreasing_list'"))
numpy_decreasing = pd.DataFrame(numpy.query("list_type == 'decreasing_list'"))
python_decreasing = pd.DataFrame(python.query("list_type == 'decreasing_list'"))
```

## Increasing lists

```python
increasing_lists = [merge_increasing,
                    quick_increasing, 
                    tim_increasing, 
                    insertion_increasing,
                    bubble_increasing,
                    python_increasing,
                    numpy_increasing
                 ]
res_increasing = pd.concat(increasing_lists)
res_increasing['milli_time'] = round(1000000*(res_increasing['time']),2)
data_wide_increasing = res_increasing.pivot('size', 'Algorithm', 'milli_time')
data_wide_increasing
```

```python
print(data_wide_increasing.to_latex())
```

## Decreasing lists

```python
decreasing_lists = [merge_decreasing,
                    quick_decreasing, 
                    tim_decreasing, 
                    insertion_decreasing,
                    bubble_decreasing,
                    python_decreasing,
                    numpy_decreasing
                 ]
res_decreasing = pd.concat(decreasing_lists)
res_decreasing['milli_time'] = round(1000000*(res_decreasing['time']),2)
data_wide_decreasing = res_decreasing.pivot('size', 'Algorithm', 'milli_time')
data_wide_decreasing
```

```python
print(data_wide_decreasing.to_latex())
```
