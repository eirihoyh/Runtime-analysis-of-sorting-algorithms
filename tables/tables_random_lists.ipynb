# Tables of Random lists

### Read and sort data by list type

```python
import pandas as pd
merge = pd.read_pickle('../data/benchmark_merge_sort')
merge['Algorithm'] = 'Merge sort'
tim = pd.read_pickle('../data/benchmark_tim_sort')
tim['Algorithm'] = 'Tim sort'
quick = pd.read_pickle('../data/benchmark_quick_sort')
quick['Algorithm'] = 'Quick sort'
insertion = pd.read_pickle('../data/benchmark_insertion_sort')
insertion['Algorithm'] = 'Insertion sort'
bubble = pd.read_pickle('../data/benchmark_bubble_sort')
bubble['Algorithm'] = 'Bubble sort'
numpy = pd.read_pickle('../data/benchmark_numpy_sort')
numpy['Algorithm'] = 'Numpy sort'
python = pd.read_pickle('../data/benchmark_python_sort')
python['Algorithm'] = 'Pyhton sort'
```

```python
merge_rand_int = pd.DataFrame(merge.query("list_type == 'random_int_list'"))
tim_rand_int = pd.DataFrame(tim.query("list_type == 'random_int_list'"))
quick_rand_int = pd.DataFrame(quick.query("list_type == 'random_int_list'"))
insertion_rand_int = pd.DataFrame(insertion.query("list_type == 'random_int_list'"))
bubble_rand_int = pd.DataFrame(bubble.query("list_type == 'random_int_list'"))
numpy_rand_int = pd.DataFrame(numpy.query("list_type == 'random_int_list'"))
python_rand_int = pd.DataFrame(python.query("list_type == 'random_int_list'"))
```

```python
merge_rand_float = pd.DataFrame(merge.query("list_type == 'random_float_list'"))
tim_rand_float = pd.DataFrame(tim.query("list_type == 'random_float_list'"))
quick_rand_float = pd.DataFrame(quick.query("list_type == 'random_float_list'"))
insertion_rand_float = pd.DataFrame(insertion.query("list_type == 'random_float_list'"))
bubble_rand_float = pd.DataFrame(bubble.query("list_type == 'random_float_list'"))
numpy_rand_float = pd.DataFrame(numpy.query("list_type == 'random_float_list'"))
python_rand_float = pd.DataFrame(python.query("list_type == 'random_float_list'"))
```

```python
merge_string = pd.DataFrame(merge.query("list_type == 'string_list'"))
tim_string = pd.DataFrame(tim.query("list_type == 'string_list'"))
quick_string = pd.DataFrame(quick.query("list_type == 'string_list'"))
insertion_string = pd.DataFrame(insertion.query("list_type == 'string_list'"))
bubble_string = pd.DataFrame(bubble.query("list_type == 'string_list'"))
numpy_string = pd.DataFrame(numpy.query("list_type == 'string_list'"))
python_string = pd.DataFrame(python.query("list_type == 'string_list'"))
```

## Random integer list

```python
rand_int_lists = [merge_rand_int, 
                  quick_rand_int, 
                  tim_rand_int, 
                  insertion_rand_int,
                  bubble_rand_int,
                  python_rand_int,
                  numpy_rand_int
                 ]
res_int = pd.concat(rand_int_lists)
res_int['milli_time'] = round(1000000*(res_int['time']), 2)
data_wide_int = res_int.pivot('size', 'Algorithm', 'milli_time')
data_wide_int
```

```python
print(data_wide_int.to_latex())
```

## Random float list

```python
rand_float_lists = [merge_rand_float, 
                    quick_rand_float, 
                    tim_rand_float, 
                    insertion_rand_float,
                    bubble_rand_float,
                    python_rand_float,
                    numpy_rand_float
                    ]
res_float = pd.concat(rand_float_lists)
res_float['milli_time'] = round(1000000*(res_float['time']), 2)
data_wide_float = res_float.pivot('size', 'Algorithm', 'milli_time')
data_wide_float
```

```python
print(data_wide_float.to_latex())
```

## String list

```python
string_lists = [merge_string, 
                quick_string, 
                tim_string, 
                insertion_string,
                bubble_string,
                python_string,
                numpy_string
                ]
res_string = pd.concat(string_lists)
res_string['milli_time'] = round(1000000*(res_string['time']),2)
data_wide_string = res_string.pivot('size', 'Algorithm', 'milli_time')
data_wide_string
```
